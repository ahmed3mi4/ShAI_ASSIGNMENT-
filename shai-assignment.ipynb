```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/salaries/Salaries.csv')
```

```python
#delete the empty columns
df.drop( ['Notes','Status'], axis=1, inplace=True )
```

```python
df.shape
```

```python
#fill the null val
df['BasePay']=df['BasePay'].fillna(df['BasePay'].mean(),inplace=False)
df['Benefits']=df['Benefits'].fillna(df['Benefits'].mean(),inplace=False)
df['OvertimePay']=df['OvertimePay'].fillna(df['OvertimePay'].mean(),inplace=False)
df['OtherPay']=df['OtherPay'].fillna(df['OtherPay'].mean(),inplace=False)
```

```python
df.isnull().sum()
```

```python
#it's time for some visuals
```

```python
#variable for only numeric data
numeric_values=df.select_dtypes(include = ['number'])
#create heatmap
sns.heatmap(data=numeric_values.corr(),cmap='Greens',annot=True)
plt.show()
```

```python
sns.relplot(data=df, x='BasePay', y='OtherPay', hue='TotalPay')
```
